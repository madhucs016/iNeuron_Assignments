```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load Data set from sklearn.datasets
iris = datasets.load_iris()
```

```python
#from PCA reduce dimension from 4 to 3 and assign Y as target
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(iris.data)
Y = iris.target
```

```python
fig = plt.figure(1,figsize = (10,12))
ax = Axes3D(fig,elev = -150,azim = 110)
ax.scatter(principalComponents[:,0],principalComponents[:,1],principalComponents[:,2],
           c=Y,cmap = plt.cm.rainbow_r)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_zlabel('Principal Component 3', fontsize = 15)
ax.set_title('component PCA', fontsize = 20)
ax.grid()
plt.show()
```
